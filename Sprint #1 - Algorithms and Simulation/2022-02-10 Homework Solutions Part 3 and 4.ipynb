```python
def factorial(N):
    value=1
    for i in range(1,N+1):
        value=value*i
        
    return value
    
    
def mysin(x,order):
    
    if order==1:
        value=x
        
    elif order==3:
        value = x - x**3/factorial(3)
        
    elif order==5:
        value = x - x**3/factorial(3) + x**5/factorial(5)
    
    return value
```

```python
def mysin(x,order):
    
    term=1
    sign=1
    
    value=0
    
    while term<=order:
        print(term,sign)

        value=value+ sign*x**term/factorial(term)
    
        term=term+2
        sign=sign*-1
        
    
    return value
```

```python
def mysin3(x,max_n):
    value=0

    for n in range(max_n+1):
        value=value+(-1)**n/factorial(2*n+1)*x**(2*n+1)
        
    return value
        
    
    
```

What is the 10001st prime number?

```python
def isprime(x):
    
    for n in range(2,x):
        if x%n==0:
            return False
        
    return True
```

```python
%%time
count=0
for x in range(2,200000):
    if isprime(x):
        count=count+1
        if count==10001:
            print(x)
            break
```

```python
def isprime2(x):
    
    if x==2:
        return True
    
    max_n=int(x/2+1)+1

    for n in range(2,max_n):
        if x%n==0:
            return False
        
    return True
```

```python
%%time
count=0
for x in range(2,200000):
    if isprime2(x):
        count=count+1
        if count==10001:
            print(x)
            break
```

```python
def isprime3(x,current_primes=[]):
    
    for n in current_primes:
        if x%n==0:
            return False
        
    return True        
        
```

```python
%%time
count=0
current_primes=[]
for x in range(2,200000):
    if isprime3(x,current_primes):
        current_primes.append(x)
        count=count+1
        if count==10001:
            print(x)
            break
```
