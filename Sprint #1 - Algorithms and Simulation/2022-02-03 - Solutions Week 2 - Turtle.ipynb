```python
def draw_square(size):
    forward(size)
    right(90)
    forward(size)
    right(90)
    forward(size)
    right(90)
    forward(size)
    right(90)    
```

```python
def move_up_and_left(distance):
    penup()
    left(90)
    forward(distance)
    left(90)
    forward(distance)
    right(180)
    pendown()    
```

```python
reset()
draw_square(50)

move_up_and_left(5)

draw_square(60)
```
