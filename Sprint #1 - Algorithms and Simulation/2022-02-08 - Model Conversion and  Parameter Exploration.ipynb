```python
# initial values of variables
x=0
v=8

# parameters
a=-10

# initial time
t=0
dt=0.1

S=Storage()

S+=t,x,v
while t<20:
    dx=v*dt
    dv=a*dt
    
    x=x+dx
    v=v+dv
    t=t+dt
    
    S+=t,x,v
    
t,x,v=S.arrays()
    
    
```

```python
plot(t,x)
xlabel('time')
ylabel('x-position')
```

```python
plot(t,v)
xlabel('time')
ylabel('speed')
```
