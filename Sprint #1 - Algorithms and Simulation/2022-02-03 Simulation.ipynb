$$
\frac{dx}{dt}=v
$$

$$
\frac{dv}{dt}=a
$$

$$
a=-10
$$


```python
S=Storage()

a=-10  # parameter

x=20  # initial conditions for each variable
v=0
t=0

S+=t,x,v
for i in range(20):
    
    # calculate change for each variable
    dx=v*dt
    dv=a*dt
    
    
    # uodate each variable
    x=x+dx
    v=v+dv
    t=t+dt
    
    # store the calculations
    S+=t,x,v

t,x,v=S.arrays()
```

```python
plot(t,x)
xlabel('time')
ylabel('position')
```

```python
plot(t,v)
xlabel('time')
ylabel('speed')
```
