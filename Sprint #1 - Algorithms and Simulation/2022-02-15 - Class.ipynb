## Exponential growth

```python
# initial values of variables
po=1
p=po
t=0

# values of parameters
a=2

dt=0.01

S=Storage()
S+=t,p

while t<3:
    
    dp=a*p*dt
    
    # update variables for a small time step dt
    
    p=p+dp
    t=t+dt
    
    S+=t,p

t,p=S.arrays()
```

```python
plot(t,p)
ylabel('population')
xlabel('time')
```

$$
\frac{dp}{dt}=a\cdot p
$$

$$
p(t) = p_o e^{at}
$$

```python
plot(t,p)

p_analysis=po*exp(a*t)
plot(t,p_analysis)
ylabel('population')
xlabel('time')
```

```python
plot(t,p,color='cyan')

p_analysis=po*exp(a*t)
plot(t,p_analysis,color='#871cc9')
ylabel('population')
xlabel('time')
```
